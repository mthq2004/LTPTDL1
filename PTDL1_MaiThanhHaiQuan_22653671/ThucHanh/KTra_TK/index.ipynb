```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc file dữ liệu
file_path = "./sales_data_sample.xlsx"
df = pd.read_excel(file_path)

# Vẽ histogram của tổng giá trị đơn hàng (SALES)
plt.figure(figsize=(10,6))
sns.histplot(df['SALES'], bins=100, kde=True, color="steelblue", edgecolor="black")

# Thiết lập tiêu đề và nhãn trục
plt.title("Phân phối tổng giá trị đơn hàng", fontsize=14)
plt.xlabel("Tổng giá trị đơn hàng (SALES)", fontsize=12)
plt.ylabel("Tần suất", fontsize=12)

# Hiển thị biểu đồ
plt.show()
```
