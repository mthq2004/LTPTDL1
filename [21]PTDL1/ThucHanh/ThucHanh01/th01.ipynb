```python
import pandas as pd
import matplotlib.pyplot as mlt
import seaborn as sns
import numpy as np
```

# 1. Đặt tên cột và đọc dữ liệu 

```python
column_name = ['Id', 'Name', 'Age', 'Weight', 'm0006', 'm0612', 'm1218', 'f0006', 'f0612', 'f1218']
df = pd.read_csv('./patient_heart_rate.csv', names=column_name)
df.head(10)
```

# 2. Tách cột “Name” thành hai cột “Firstname”và “Lastname”, 

```python
df[['Firstname', 'Lastname']] = df['Name'].str.split(expand=True)
df = df.drop('Name', axis= 1)
```

##### df['Name']: Truy cập cột có tên là Name trong DataFrame df.
- str.split(): Sử dụng phương thức split() để chia chuỗi trong cột Name dựa trên khoảng trắng mặc định (hoặc có thể chỉ định ký tự phân tách). Chuỗi sẽ được chia thành hai phần: họ và tên.
  #
- expand=True: Tùy chọn expand=True sẽ trả về kết quả chia tách dưới dạng hai cột thay vì một chuỗi các danh sách.
  #
- df[['Firstname', 'Lastname']]: Kết quả sau khi chia tách sẽ được gán vào hai cột mới trong DataFrame, Firstname (tên) và Lastname (họ).
  #
- df.drop('Name', axis=1): Lệnh này xóa cột Name khỏi DataFrame df. axis=1 chỉ định rằng cột cần được xóa (nếu là hàng thì axis=0).

```python
df.head(50)
```

# 3. Đổi đơn vị Weight thành kg 

```python
weight = df['Weight']

for i in range(0, len(weight)):
    x = str(weight[i])
    if 'lbs' in x[-3:]:
        x = x[:-3:]
        float_x = float(x)
        y = int(float_x / 2.2)
        # y = str(y)
        weight[i] = y
    if 'kgs' in x[-3:]:
        x = x[:-3:]
        float_x = float(x)
        weight[i] = x
        
```

# 4. Đổi tên cột Weight thành cột Weight_kgs

```python
df.rename(columns={'Weight': 'Weight_kgs'}, inplace= True)
```

```python
df
```

# 5. Xóa Nan

```python
df.dropna(how='all', inplace=True)
df
```

# 6. Xóa trùng lặp

```python
df.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight_kgs'])
df
```

# 7. Xóa dữ liệu 

```python
df.Firstname.replace({r'[^\x00-\x7F]+':''}, regex = True, inplace=True)
df.Lastname.replace({r'[^\x00-\x7F]+':''}, regex = True, inplace=True)
df
# Xuất hiện dữ liệu bị ảnh hưởng bởi lỗi non-ASCII, không định dạng ASCII. 
# Giải pháp: Tùy vào nghiệp vụ ta có thể: xóa dữ liệu tại đó, 
# thay thế bằng dữ liệu khác hoặc thay bằng việc đánh dấu bằng một kí tự khác (ví dụ: ‘warning’)

```

```python
# {r'[^\x00-\x7F]+':''}:
# r'[^\x00-\x7F]+': Đây là một biểu thức chính quy (regular expression, hay regex).
# r'': r trước dấu ngoặc kép biểu thị rằng chuỗi sau nó là một raw string (chuỗi thô), để Python hiểu rõ biểu thức regex.
# [^\x00-\x7F]+: Đây là phần quan trọng của biểu thức chính quy:
# \x00-\x7F: Phạm vi ký tự từ \x00 đến \x7F đại diện cho tất cả các ký tự ASCII chuẩn (từ mã 0 đến mã 127).
# ^: Dấu ^ trong dấu ngoặc vuông ([]) nghĩa là phủ định, tức là tìm các ký tự không thuộc phạm vi ASCII (các ký tự ngoài mã từ \x00 đến \x7F).
# +: Dấu cộng + nghĩa là tìm một hoặc nhiều ký tự không phải ASCII.
# '': Thay thế bất kỳ chuỗi ký tự nào không phải ASCII bằng một chuỗi rỗng, tức là xóa chúng.
```

# 8. Xử lí số liệu


#### A. Thay giá trị thiếu của tuổi bằng giá trị yếu vị.

```python
df['Age'].fillna(df['Age'].mode()[0], inplace=True)
```

#### B. Thay giá trị bị thiếu của cân nặng làm trung vị 

```python
# Ép kiểu dữ liệu thành FLoat 
df['Weight_kgs'] = pd.to_numeric(df['Weight_kgs'], errors='coerce')

df['Weight_kgs'].fillna(df['Weight_kgs'].median(), inplace=True)
```

# 9. Một cột chứa quá nhiều thông tincần được phân rã”, như trong bài toán này ta thấy header “m0006” chứa các nội dung bao gồm: m →male, 0006~ 00-06(lần đochỉsốnhịp timtừ00h-đến 06h). Còn giá trịthì là kết quảđo nhịp tim

#### B1: Tạo melt dữ liệu để có cột gender_time 

```python
df = pd.melt(df, id_vars=['Id','Age', 'Weight_kgs','Firstname', 'Lastname'], value_name='PulseRate', var_name='gender_time').sort_values(['Id', 'Weight_kgs','Firstname', 'Lastname'])
```

B2: Tạo data frame tạm là kết quảcủa việc tách cột gender_tim

```python
df_temp = df['gender_time'].str.extract("(\D)(\d+)(\d{2})", expand=True)
df_temp
```

B3: Đặt tên cho data frame tạm vào data frame  ban đầu 

```python
df_temp.columns = ['Gender', 'Lower_hour', 'Upper_hour']
```

B4: Nối data frame tạm vào data frame ban đầu 

```python
df_temp = pd.concat([df,  df_temp], axis=1)
```

# 10. Loại bỏ hết các dòng dữliệu thừa là những dòng có phần PulseRate có dấu

```python
df = df.reset_index()
```
