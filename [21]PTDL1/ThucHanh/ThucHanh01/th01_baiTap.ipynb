# Phân tích dữ liệu 

```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('patient_heart_rate_clean.csv')
df
```

#### 1.Vẽ biểu đồ Pie đểthể hiện tỉ lệ bệnh nhân nam và nữ


```python
import matplotlib.pyplot as plt

gender_count = df['Gender'].value_counts()
labels = gender_count.index
sizes = gender_count.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Tỉ lệ bệnh nhân nam và nữ')
plt.show()
```

#### 2.Hãy tạo cột Fullname = Firstname+’‘+Lastname


```python
df['Fullname'] = df['Firstname'] + ' ' + df['Lastname']
df
```

#### 3.Vẽ biểu đồ cột so sánh độtuổi giữa cáctênbệnh nhân



```python
# 3. Vẽ biểu đồ cột so sánh độ tuổi giữa các tên bệnh nhân
plt.figure(figsize=(25,10))
sns.barplot(x='Fullname', y='Age', data=df, errorbar=None)
plt.show()
```

#### 4.Vẽ biểu đồ cột so sánh nhịp tim trung bình giữa cáctên bệnh nhân


```python
# 4. Vẽ biểu đồ cột so sánh nhịp tim trung bình giữa các tên bệnh nhân
PulseRate_avg = df.groupby('Fullname')['PulseRate'].mean().reset_index()
PulseRate_avg
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x='Fullname', y='PulseRate', data=PulseRate_avg)
plt.show()
```

#### 5.Vẽ biểu đồcột so sánh nhịp timtrung bình theo từng nhóm giờ 00-06, 06-12,12-18 có phân loại theo giới tính


```python
# 5. Vẽ biểu đồ cột so sánh nhịp tim trung bình theo từng nhóm giờ 00-06, 06-12,12-18 có 
# phân loại theo giới tính
df['pulse_time'] = df['Lower_hour'].astype(str) + '-' + df['Upper_hour'].astype(str)
```

```python
pulse_time_gender = df.groupby(['pulse_time', 'Gender'])['PulseRate'].mean().reset_index()
pulse_time_gender
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x='pulse_time', y='PulseRate', hue='Gender', data=pulse_time_gender)
plt.show()
```

#### 6.Vẽ biểu đồphân tán thểhiện tương quan giữa trung bình độtuổi và trung bình nhịp tim


```python
# 6. Vẽ biểu đồ phân tán thể hiện tương quan giữa trung bình độ tuổi và trung bình nhịp tim
pulse_age = df.groupby(['Age'])['PulseRate'].mean().reset_index()
pulse_age
```

```python
sns.lmplot(data=pulse_age, x='Age', y='PulseRate', fit_reg=True)
plt.show()
```

#### 7.Vẽ biểu đồđường thể hiện sự thay đổi của nhịp tim trung bình các bệnh nhân theo thứ tự khung giờ 00-06, 06-12,12-18 theo từng nhóm giới tính.

```python
sns.lmplot(data=pulse_age, x='Age', y='PulseRate', fit_reg=True)
plt.show()
```

```python
pulse_time_gender = df.groupby(['pulse_time', 'Gender'])['PulseRate'].mean().reset_index()
pulse_time_gender
```

```python
sns.lineplot(x='pulse_time', y='PulseRate', hue='Gender', 
             data=pulse_time_gender, estimator=np.max)
plt.show()
```
