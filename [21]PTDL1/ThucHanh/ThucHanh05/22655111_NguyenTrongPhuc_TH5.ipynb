15. Dựa vào confusion matrix, hãy cho biết giá trị quan trọng nhất trong bài toán dự báo này.
Trong bài toán dự báo bệnh này, False Negative (FN = 9) là giá trị quan trọng nhất. Đây là số lượng bệnh nhân thực sự có bệnh nhưng bị dự đoán sai là không có bệnh. Điều này có thể gây hậu quả nghiêm trọng vì bệnh nhân có thể không nhận được điều trị cần thiết.

Giảm FN là rất quan trọng để đảm bảo bệnh nhân có bệnh được chẩn đoán và điều trị đúng cách. Tuy nhiên, bạn cũng cần chú ý giảm False Positive (FP = 10) để tránh các chẩn đoán nhầm lẫn không cần thiết.

Tổng kết:

FN (9) cần được giảm càng nhiều càng tốt để tránh bỏ sót các bệnh nhân có bệnh.
FP (10) cũng cần được giảm nhưng mức độ quan trọng thấp hơn so với FN trong các bài toán y tế.

Đại lượng quan trọng nhất:
Trong bài toán dự báo bệnh, Recall thường quan trọng nhất vì:

Recall đo lường khả năng phát hiện đúng các bệnh nhân có bệnh (True Positive) và giảm thiểu các trường hợp âm tính giả (False Negative).
False Negative (FN) có thể gây ra hậu quả nghiêm trọng nếu bệnh nhân thực sự có bệnh nhưng không được chẩn đoán đúng.
F1-score cũng có ý nghĩa nếu bạn muốn cân bằng giữa Precision và Recall, nhưng Recall vẫn là đại lượng chính yếu cần ưu tiên tối ưu trong bài toán y tế.

```python
# 19.  Hãy in ra màn hình accuracy,  confusion matrix, [tn,  fp,  fn,  tp],  [TPR, FNR, FPR, TNR],
# [precision, recal, F1] và đồ thị  AUC & ROC với ngưỡng xác xuất phân lớp là 0.65. Đưa ra 
# kết luận  khi thiết lập threshold tăng hoặc threshold giảm sẽ  tác động như thế  nào đến kết 
# quả đánh giá mô hình.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    roc_curve,
    auc,
    precision_score,
    recall_score,
    f1_score
)

# Giả sử y_test là nhãn thực và y_proba là xác suất dự đoán từ mô hình
y_test = [0, 1, 0, 1, 1]  # Thay thế bằng nhãn thực tế của bạn
y_proba = [0.1, 0.9, 0.8, 0.6, 0.4]  # Thay thế bằng xác suất dự đoán của mô hình

# Thiết lập ngưỡng
threshold = 0.65
y_pred = (np.array(y_proba) >= threshold).astype(int)

# Tính toán accuracy
accuracy = accuracy_score(y_test, y_pred)

# Tính toán confusion matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

# Tính toán TPR, FNR, FPR, TNR
TPR = tp / (tp + fn)  # True Positive Rate (Recall)
FNR = fn / (tp + fn)  # False Negative Rate
FPR = fp / (fp + tn)  # False Positive Rate
TNR = tn / (tn + fp)  # True Negative Rate

# Tính toán precision, recall, F1
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# In ra kết quả
print(f"Accuracy: {accuracy:.2f}")
print(f"Confusion Matrix: [TN: {tn}, FP: {fp}, FN: {fn}, TP: {tp}]")
print(f"[TPR: {TPR:.2f}, FNR: {FNR:.2f}, FPR: {FPR:.2f}, TNR: {TNR:.2f}]")
print(f"[Precision: {precision:.2f}, Recall: {recall:.2f}, F1: {f1:.2f}]")

# Đồ thị ROC và AUC
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd

# Giả sử bạn đã huấn luyện mô hình knn_best ở đây

# Hàm để dự đoán bệnh tim
def predict_heart_disease(model, sample_input, threshold=0.5):
    # Dự đoán xác suất
    probability = model.predict_proba(sample_input)[:, 1]  # Lấy xác suất cho lớp 1 (có bệnh tim)
    
    # Dựa trên ngưỡng để đưa ra dự đoán
    if probability[0] >= threshold:
        return 1, probability[0]  # Bệnh tim
    else:
        return 0, probability[0]  # Không có bệnh tim

def main():
    print("Nhập các đặc trưng đầu vào cho bệnh tim:")
    
    # Nhập các đặc trưng từ người dùng
    age = float(input("Tuổi (age): "))
    sex = int(input("Giới tính (sex) (0: nữ, 1: nam): "))
    cp = int(input("Loại đau ngực (cp) (0-3): "))
    trestbps = float(input("Huyết áp (trestbps): "))
    chol = float(input("Mỡ máu (chol): "))
    fbs = int(input("Đường huyết lúc đói > 120 mg/dl (fbs) (0: không, 1: có): "))
    restecg = int(input("Điện tâm đồ (restecg) (0-2): "))
    thalach = float(input("Nhịp tim tối đa (thalach): "))
    exang = int(input("Cơn đau thắt ngực do gắng sức (exang) (0: không, 1: có): "))
    oldpeak = float(input("Giảm ST (oldpeak): "))
    slope = int(input("Độ dốc ST (slope) (0-2): "))
    ca = int(input("Số mạch máu (ca) (0-3): "))
    thal = int(input("Thalassemia (thal) (0-3): "))

    # Tạo mảng đặc trưng từ các đầu vào
    sample_input = np.array([[age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal]])
    
    # Nhập ngưỡng xác suất
    threshold_input = input("Nhập ngưỡng xác suất (mặc định là 0.5): ")
    
    # Nếu người dùng không nhập ngưỡng, sử dụng ngưỡng mặc định
    if threshold_input.strip() == "":
        threshold = 0.5
    else:
        threshold = float(threshold_input)

    # Dự đoán bệnh tim
    prediction, probability = predict_heart_disease(knn, sample_input, threshold)

    # In ra kết quả
    if prediction == 1:
        print(f"Có bệnh tim. Xác suất: {probability:.2f}")
    else:
        print(f"Không có bệnh tim. Xác suất: {probability:.2f}")

if __name__ == "__main__":
    main()
```
