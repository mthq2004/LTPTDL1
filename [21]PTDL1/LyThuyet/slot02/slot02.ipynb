```python
# Thay thế dữ liệu thiếu
df['DT'].fillna('KINH', inplace=True)
df['DT']
```

```python
#Điền giá trị thiếu phần DiemThi1 bằng điểm trung bình
df['M1'].fillna(df['M1'].mean(), inplace=True)
df['M1']

```

```python
#Điền giá trị thiếu phần DiemThi2 bằng trung vị
df['M2'].fillna(df['M2'].median(), inplace=True)
```

```python

#Điền giá trị thiếu phần diemthi3 bằng 0 nếu có
df['M3'].fillna(0, inplace=True)
```

df

####  5. Tạo cột 

```python
# Tạo cột TBM (Điểm trung bình)
df['TBM'] = (df['M1'] + df['M2'] + df['M3']) / 3
df 
```

```python
# Gán giá trị cho cột XL dựa trên điểm trung bình
df.loc[df['TBM'] < 5.0, 'XL'] = 'FAIL'
df

```

```python
df.loc[(df['TBM'] >= 5.0) & (df['TBM'] < 7.0), 'XL'] = 'FAIR'
df.loc[(df['TBM'] >= 7.0) & (df['TBM'] < 9.0), 'XL'] = 'GOOD'
df.loc[df['TBM'] >= 9.0, 'XL'] = 'EXCEL'
df
```

```python
# Tạo cột NHOM dựa trên ánh xạ từ cột KT
# Tạo một cột NHOM dựa trên cột KT 
dict_map = {'A1': 'G1', 'C': 'G3', 'D1': 'G3', 'A': 'G1', 'B': 'G2'}
# Tạo cột NHOM  , lấy các ảnh xạ được lưu vào cột NHOM 
df['NHOM'] = df['KT'].map(dict_map) # Sử dụng hàm map() để thay thế giá trị của cột KT theo từ điển ánh xạ dict_map
df
```

```python
# Tính cột PLUS dựa trên điều kiện
def fplus(x, y):
    if (x == 'G1' or x == 'G2') and (y >= 4.0):
        return 1.0
    else:
        return 0.0
# list gọi hàm và 2 cột NHOM và TBM
# lệnh list(map(...)) được sử dụng để tính toán giá trị cho cột PLUS dựa trên hàm fplus và hai cột NHOM và TBM trong DataFrame df.
df['PLUS'] = list(map(fplus, df['NHOM'], df['TBM']))
df
```

#### 8. Pie

- Biểu đồ tròn phù hợp khi có ít danh mục (thường dưới 6-7 danh mục) để tránh việc biểu đồ bị rối.
- Thường dùng cho biến định tính 
- labels: Danh sách chứa tên của các nhóm, trong trường hợp này là các giá trị từ cột KT.
- data: Danh sách chứa số lượng (đếm) của từng nhóm, được tính từ df.groupby(['KT'])['KT'].agg(['count']).
- colors: Danh sách chứa màu sắc cho từng phần của biểu đồ. Bạn có thể sử dụng palette từ seaborn như trong ví dụ.

```python
# Biểu đồ tròn phân bố số lượng nhóm KT
# median , 
gb = df.groupby(['KT'])['KT'].agg(['count'])
labels = gb.index
data = list(gb['count'])
colors = sns.color_palette('pastel')
plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
# Biểu đồ tròn phân bố số lượng PLUS theo KV
gb = df.groupby(['KV'])['PLUS'].agg(['sum'])
labels = gb.index
data = list(gb['sum'])
colors = sns.color_palette('pastel')
plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.show()
```
