#### 3. Tóm tắt dữ liệu (Data Summary):

#
- df.aggregate({'M1':['min','max'],'M2':['mean','sum']}): Tính toán giá trị nhỏ nhất, lớn nhất của cột M1, và giá trị trung bình, tổng của cột M2.
#
- df.groupby(['KV'])['KV'].agg(['count']): Đếm số lượng các giá trị cho mỗi khu vực KV.
#
- df.groupby(['KV'])['M1'].agg(['count','min','mean','std','max','sum']): Tính số lượng, giá trị nhỏ nhất, trung bình, độ lệch chuẩn, giá trị lớn nhất và tổng của cột M1 theo từng KV.
#
- df.groupby(['KV'])[['M1','M2']].agg(['min','mean']): Tính giá trị nhỏ nhất và trung bình của M1 và M2 theo từng KV.
#
- df.groupby(['KV','KT'])[['M1','M2']].agg(['min','max']): Tính giá trị nhỏ nhất và lớn nhất của M1 và M2 theo từng cặp KV và KT.
#
- df_gr = df.groupby(['KV'],as_index=False).count(): Nhóm theo KV và đếm số lượng các giá trị, lưu vào df_gr.
#
- pd.pivot_table(df,values=['M1','M2'], columns='KV', aggfunc=['min','mean','max']): Tạo bảng pivot với giá trị nhỏ nhất, trung bình, và lớn nhất của M1 và M2 theo KV.
#
- pd.pivot_table(newdf, values=['M1','M2'], columns=['KV','KT'], aggfunc=['min','mean','max']): Tạo bảng pivot với newdf để tính giá trị nhỏ nhất, trung bình và lớn nhất của M1, M2 theo KV và KT.

#### Tạo bảng pivot 

Bảng pivot (pivot table) là một công cụ phân tích dữ liệu mạnh mẽ trong pandas, cho phép bạn tóm tắt và phân tích dữ liệu một cách linh hoạt. Dưới đây là một số mục đích và lợi ích của việc sử dụng bảng pivot

```python
pd.pivot_table(df,values=['M1','M2'], columns='KV', aggfunc=['min','mean','max'])
# Tạo bảng pivot với giá trị nhỏ nhất, trung bình, và lớn nhất của M1 và M2 theo KV.
```

```python
pd.pivot_table(newdf, values=['M1','M2'], columns=['KV','KT'], aggfunc=['min','mean','max'])
# Tạo bảng pivot với newdf để tính giá trị nhỏ nhất, trung bình và lớn nhất của M1, M2 theo KV và KT.
```

```python
pd.pivot_table(
    df,                 # DataFrame hoặc array-like để tạo pivot table
    values=['M1'],          # Cột mà bạn muốn tính toán trên
    columns=['KT'],         # Cột hoặc danh sách cột để nhóm (tương ứng với cột)
    aggfunc=['mean'],       # Hàm tổng hợp (aggregate function) mặc định là 'mean', có thể dùng 'sum', 'count', 'min', 'max', v.v.
   
)
```
