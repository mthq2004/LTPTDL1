### Thao Tác với DataFrame (Data Frame Manipulation)
#### 1. Truy Cập và Chọn Lọc Dữ Liệu:

```python
# Truy cập cột M1.
df['M1']
```

```python
# Truy cập cột M1 dưới dạng DataFrame.
df[['M1']]
```

```python
# Tạo DataFrame mới newdf chỉ gồm các cột M1, M2, KV, và KT.
newdf = df[['M1','M2','KV','KT']]
newdf
```

```python
#  Lấy dữ liệu từ dòng thứ 2 đến dòng thứ 96.
df[2:97]
```

```python
# Lấy dữ liệu tại dòng có chỉ số 92.
df.loc[92]
```

```python
# Lấy dữ liệu từ dòng 4 đến dòng 10.
df.loc[4:10]
```

```python
#  Lấy dữ liệu từ dòng 5 đến dòng 9, chỉ bao gồm các cột M1 và KT.
df.loc[5:9, ['M1', 'KT']]
```

#### 2. Sắp Xếp Dữ Liệu:



```python
# Sắp xếp dữ liệu theo cột M1 theo thứ tự tăng dần.
df.sort_values(by='M1', ascending=True)
```

```python
# Sắp xếp theo M1 tăng dần và M2 giảm dần
df.sort_values(by=['M1', 'M2'], ascending=[True, False])
```

#### 3. Lọc dữ liệu 

```python
# Lọc ra các dòng có giá trị M1 lớn hơn hoặc bằng 5.
df[df['M1'] >= 5] 
```

```python
# Lọc ra các dòng có M1 lớn hơn 6 và M2 lớn hơn 5.
df[(df['M1'] > 6) & (df['M2'] > 5)]
```
