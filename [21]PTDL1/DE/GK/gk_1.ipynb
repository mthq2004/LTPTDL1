#### Tìm hiểu và giải thích về bộ dữ liệu. Cho biết biến nào là định tính, biến nào là định lượng, kiểu dữ liệu và thang đo cho mỗi thuộc tính.

```python
import pandas as pd

def phan_loai_bien_va_thang_do(df):
    # Tạo danh sách để lưu biến định lượng và biến định tính
    dinhluong_column = []
    dinhtinh_column = []
    
    # Duyệt qua từng cột trong DataFrame
    for col in df.columns:
        dtype = df[col].dtype
        
        if dtype in ['int64', 'float64']:
            dinhluong_column.append(col)  # Biến định lượng
            # Xác định thang đo cho biến định lượng
            scale = "Ratio" if dtype == 'float64' else "Ordinal"
            print(f"Biến '{col}' là định lượng ({dtype}) - Thang đo: {scale}")
        else:
            dinhtinh_column.append(col)    # Biến định tính
            # Xác định thang đo cho biến định tính
            scale = "Nominal"
            print(f"Biến '{col}' là định tính ({dtype}) - Thang đo: {scale}")

    # In danh sách biến định lượng và định tính
    print(f"\nBiến định lượng: {dinhluong_column}")
    print(f"Biến định tính: {dinhtinh_column}")

# Ví dụ sử dụng hàm
# Giả sử df là DataFrame đã được định nghĩa và có dữ liệu
# df = pd.read_csv('duong_dan_toi_file.csv')
phan_loai_bien_va_thang_do(df)
```

#### PHẦN 3: XỬ LÝ DỮ LIỆU – TRỰC QUAN HOÁ


#### 1. Giả sử bạn muốn tạo một biến mới có tên là "PerformanceScore" bằng cách tính trung bình của "ProfessionalScore" và "TimeToStudy". Dưới đây là mã để thực hiện điều này.

```python
import pandas as pd

# Giả sử df là DataFrame bạn đang sử dụng

# Hàm tính PerformanceScore
def calculate_performance_score(row):
    return (row['ProfessionalScore'] + row['TimeToStudy']) / 2

# Tạo cột mới 'PerformanceScore'
df['PerformanceScore'] = df.apply(calculate_performance_score, axis=1)

# Hiển thị dữ liệu sau khi thêm cột mới
print("Dữ liệu sau khi thêm cột PerformanceScore:")
df[['ProfessionalScore', 'TimeToStudy', 'PerformanceScore']]
```

#### 7. Boxplot và xử lý ngoại lệ: Ví dụ: Boxplot tìm ngoại lệ của ExpectedSalary và loại bỏ ngoại lệ:



```python
sns.boxplot(y='ExpectedSalary', data=merged_df)
plt.show()

Q1 = merged_df['ExpectedSalary'].quantile(0.25)
Q3 = merged_df['ExpectedSalary'].quantile(0.75)
IQR = Q3 - Q1
data_no_outliers = merged_df[(merged_df['ExpectedSalary'] >= (Q1 - 1.5 * IQR)) & (merged_df['ExpectedSalary'] <= (Q3 + 1.5 * IQR))]
print(data_no_outliers)
```

#### PHẦN 4: THỐNG KÊ SUY DIỄN


#### 1. T-Test cho biến định lượng: Ví dụ: Kiểm định trung bình của ExpectedSalary:



```python
from scipy import stats
t_stat, p_val = stats.ttest_1samp(merged_df['ExpectedSalary'], popmean=500)
print(f"T-Test: t-stat={t_stat},  p-value={p_val}")
```

#### 2. Z-Test giữa 2 biến định lượng: Ví dụ: Kiểm tra giữa TimeToPlay và TimeToStudy:



```python
from statsmodels.stats.weightstats import ztest
z_stat, p_val = ztest(merged_df['TimeToPlay'], merged_df['TimeToStudy'])
print(f"Z-Test: z-stat={z_stat}, p-value={p_val}")
```

#### 3. Kiểm định tương quan Fisher: Ví dụ: Tương quan giữa TimeToPlay và ExpectedSalary:



```python
correlation, p_value = stats.pearsonr(merged_df['TimeToPlay'], merged_df['ExpectedSalary'])
print(f"Tương quan Fisher: correlation={correlation}, p-value={p_value}")
```
